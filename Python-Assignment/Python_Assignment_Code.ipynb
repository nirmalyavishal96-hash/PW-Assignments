Task 2: Mutable Object Example

```python
my_list = [1, 2, 3]

print("Original List:", my_list)

my_list.append(4)

print("Modified List:", my_list)
```

Output

Original List: [1,2,3]    
Modified List: [1,2,3,4]

Task 3: Immutable Object Example

```python
x = 10

print("Original value:", x)

x = x + 5

print("Modified value:", x)
```

Task 4: Operators Example

```python
a = 10
b = 5

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Comparison:", a > b)
```

Task 5: Type Casting Example

```python
num_str = "20"

num_int = int(num_str)

print("String value:", num_str)
print("Converted Integer:", num_int)
```

Task 6: Conditional Statement

```python
number = 10

if number % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

Task 7: For Loop Example and While Loop Example

```python
for i in range(1,6):
    print(i)
```

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```
