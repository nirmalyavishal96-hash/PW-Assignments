# Task 1 – Basic Git Workflow

## Objective

Create a folder, add files, stage selected files, commit them, and push the changes to GitHub.

---

## Step 1: Create a New Folder

```bash
mkdir git-assignment1
cd git-assignment1
```

---

## Step 2: Initialize Git Repository

```bash
git init
```

---

## Step 3: Create Required Files

```bash
touch Code.txt Log.txt Output.txt
```

Verify files:

```bash
ls
```

Expected Output:

```
Code.txt
Log.txt
Output.txt
```

---

## Step 4: Stage Code.txt and Output.txt

```bash
git add Code.txt Output.txt
```

Check staged files:

```bash
git status
```

---

## Step 5: Commit the Files

```bash
git commit -m "Added Code.txt and Output.txt files"
```

---

## Step 6: Push to GitHub

Add remote repository

```bash
git remote add origin https://github.com/<username>/<repository>.git
```

Push changes

```bash
git branch -M main
git push -u origin main
```

# Task 2 – Git Branching and Stashing

## Objective

Practice branch creation, stashing, switching branches, and committing changes.

---

## Step 1: Create Working Directory

```bash
mkdir git-assignment2
cd git-assignment2
git init
```

---

## Step 2: Create Files in Master Branch

```bash
touch feature1.txt feature2.txt
```

Stage and commit:

```bash
git add .
git commit -m "Added feature1.txt and feature2.txt"
```

---

## Step 3: Create Branches

```bash
git branch develop
git branch feature1
git branch feature2
```

---

## Step 4: Switch to Develop Branch

```bash
git checkout develop
```

---

## Step 5: Create develop.txt (Do Not Commit)

```bash
touch develop.txt
```

Check status:

```bash
git status
```

---

## Step 6: Stash the File

```bash
git stash
```

---

## Step 7: Switch to feature1 Branch

```bash
git checkout feature1
```

---

## Step 8: Create new.txt File

```bash
touch new.txt
```

Stage and commit:

```bash
git add new.txt
git commit -m "Added new.txt in feature1 branch"
```

---

## Step 9: Return to Develop Branch

```bash
git checkout develop
```

---

## Step 10: Restore Stashed File

```bash
git stash pop
```

---

## Step 11: Commit develop.txt

```bash
git add develop.txt
git commit -m "Added develop.txt after stash"
```

---

# Task 5 – GitFlow Workflow

## Objective

Implement a GitFlow architecture and simulate feature development and hotfix deployment.

---

## Step 1: Initialize Git Repository

```bash
mkdir gitflow-assignment
cd gitflow-assignment
git init
```

---

## Step 2: Create GitFlow Branch Structure

```bash
git branch develop
git branch feature-login
git branch release
git branch hotfix
```

---

## Step 3: Start Feature Development

```bash
git checkout feature-login
touch feature.txt
git add feature.txt
git commit -m "Added feature implementation"
```

---

## Step 4: Merge Feature Branch into Develop

```bash
git checkout develop
git merge feature-login
```

---

## Step 5: Merge Develop into Master

```bash
git checkout master
git merge develop
```

---

## Step 6: Create Hotfix for Urgent Issue

```bash
git checkout hotfix
touch urgent.txt
git add urgent.txt
git commit -m "Hotfix: urgent production fix"
```

---

## Step 7: Merge Hotfix into Master

```bash
git checkout master
git merge hotfix
```
