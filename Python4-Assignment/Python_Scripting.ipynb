1. Write Python scripts for basic file operations and data processing

Python provides built-in functions to perform file handling operations such as reading, writing, and processing data from files.

Example: Writing and Reading a File

```python
# Writing data to a file
file = open("sample.txt", "w")
file.write("Hello, this is a Python scripting example.\n")
file.write("File operations are useful for data processing.")
file.close()
```

```python
# Reading data from the file
file = open("sample.txt", "r")
content = file.read()
print(content)
file.close()
```

Example: Data Processing from File

```python
file = open("numbers.txt", "w")
file.write("10\n20\n30\n40\n50")
file.close()

file = open("numbers.txt", "r")
numbers = file.readlines()

total = 0

for num in numbers:
    total += int(num)

print("Sum of numbers:", total)

file.close()
```

2. Develop a simple web scraper to extract data from a website

Web scraping allows us to extract information from websites automatically.
Python libraries such as requests and BeautifulSoup are commonly used.

Example: Simple Web Scraper

```python
import requests
from bs4 import BeautifulSoup

url = "https://example.com"

response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

print("Page Title:", soup.title.text)
```

Example: Extract All Links from a Web Page

```python
import requests
from bs4 import BeautifulSoup

url = "https://example.com"

response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

links = soup.find_all("a")

for link in links:
    print(link.get("href"))
```
