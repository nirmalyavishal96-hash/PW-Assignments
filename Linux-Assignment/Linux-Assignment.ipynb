# 1. What is Linux?

Linux is an **open-source operating system** based on the Unix architecture. It manages computer hardware resources and provides services to application software.

Linux was created by **Linus Torvalds in 1991** and is widely used in:

* Servers
* Cloud platforms
* Supercomputers
* Embedded systems
* DevOps environments

### Popular Linux Distributions

* Ubuntu
* Red Hat Enterprise Linux
* CentOS
* Debian

# 2. Difference Between Hard Link and Soft Link

| Feature                     | Hard Link                          | Soft Link (Symbolic Link) |
| --------------------------- | ---------------------------------- | ------------------------- |
| Definition                  | Direct reference to the file inode | Pointer to the file path  |
| Inode                       | Same inode as original file        | Different inode           |
| File System                 | Cannot cross file systems          | Can cross file systems    |
| Directory Support           | Cannot link directories            | Can link directories      |
| If original file is deleted | File still accessible              | Link becomes broken       |

### Example Commands

Create a Hard Link

```bash
ln file1.txt hardlink.txt
```

Create a Soft Link

```bash
ln -s file1.txt softlink.txt
```

# 3. What is a Kernel in Linux?

The **Kernel** is the **core component of the Linux operating system**. It acts as an interface between **hardware and software**.

### Functions of the Kernel

* Process management
* Memory management
* Device driver management
* File system management
* Security and access control

Without the kernel, applications cannot communicate with the hardware.

# 4. Creating a User Account in Linux

In Linux, a user account can be created using the `useradd` command.

### Command

```bash
sudo useradd username
```

### Example

```bash
sudo useradd p1
```

### Set Password for User

```bash
sudo passwd p1
```

### Verify User Creation

```bash
cat /etc/passwd | grep p1
```

# 5. What is the `grep` Command in Linux?

The **grep command** is used to search for specific text patterns inside files.

### Syntax

```bash
grep "pattern" filename
```

### Example

```bash
grep "error" logfile.txt
```

### Common Options

| Option | Description       |
| ------ | ----------------- |
| -i     | Ignore case       |
| -n     | Show line numbers |
| -r     | Recursive search  |

Example:

```bash
grep -i "linux" file.txt
```

# 6. User and Group Management Task

## Step 1: Create User p1

```bash
sudo useradd p1
sudo passwd p1
```

---

## Step 2: Create Groups g1, g2, g3

```bash
sudo groupadd g1
sudo groupadd g2
sudo groupadd g3
```

---

## Step 3: Add User p1 to All Groups

```bash
sudo usermod -aG g1,g2,g3 p1
```

Verify group membership

```bash
groups p1
```

---

## Step 4: Set Primary Group g1

This ensures that whenever user **p1 creates a file**, the group ownership automatically becomes **g1**.

```bash
sudo usermod -g g1 p1
```

Verification

```bash
id p1
```

# 7. Directory Permission Task

## Step 1: Create Directory `/tmp/bg`

```bash
sudo mkdir /tmp/bg
```

Create files inside the directory

```bash
sudo touch /tmp/bg/file1
sudo touch /tmp/bg/file2
```

---

## Step 2: Change Ownership to User `abhi`

```bash
sudo chown abhi /tmp/bg
```

Verify ownership

```bash
ls -ld /tmp/bg
```

---

## Step 3: Assign Proper Permissions

```bash
sudo chmod 770 /tmp/bg
```

### Permission Explanation

| Permission | Meaning              |
| ---------- | -------------------- |
| Owner      | Read, Write, Execute |
| Group      | Read, Write, Execute |
| Others     | No access            |

Now user **abhi** can:

* Create files
* Delete files
* Modify files
* Add content to files

# 8. Identifying and Terminating High CPU Processes

Sometimes a process may consume excessive CPU resources. Linux provides tools to identify and terminate such processes.

---

## Step 1: Identify High CPU Usage

Use the `top` command:

```bash
top
```

This command shows:

* Running processes
* CPU usage
* Memory usage
* Process IDs

---

## Step 2: Sort Processes by CPU Usage

```bash
ps aux --sort=-%cpu
```

Example Output

```
PID   USER   %CPU   COMMAND
2345  root   95.2   java
```

---

## Step 3: Terminate the Process

Kill the process using its PID.

```bash
kill 2345
```

If the process does not terminate:

```bash
kill -9 2345
```

`-9` sends a **SIGKILL signal**, forcefully stopping the process.
