# Task 1 – File Management Script

## Objective

Create a script that creates a backup directory and copies `.txt` files into it while appending the current date and time to the filenames.

---

## Script: file_backup.sh

```bash
#!/bin/bash

backup_dir="$HOME/backup"

# Create backup directory if it does not exist
mkdir -p "$backup_dir"

# Loop through all txt files
for file in *.txt
do
    timestamp=$(date +"%Y%m%d_%H%M%S")
    filename=$(basename "$file" .txt)
    cp "$file" "$backup_dir/${filename}_${timestamp}.txt"
done

echo "Backup completed successfully."
```

### Run the Script

```bash
chmod +x file_backup.sh
./file_backup.sh
```

# Task 2 – System Health Check

## Objective

Create a script that monitors CPU and memory usage and logs system health status.

---

## Script: system_health.sh

```bash
#!/bin/bash

logfile="system_health.log"

cpu_usage=$(top -bn1 | grep "Cpu(s)" | awk '{print $2 + $4}')
memory_usage=$(free | awk '/Mem:/ {printf("%.2f"), $3/$2 * 100}')

echo "System Health Check - $(date)" >> $logfile
echo "CPU Usage: $cpu_usage%" >> $logfile
echo "Memory Usage: $memory_usage%" >> $logfile

if (( $(echo "$cpu_usage > 80" | bc -l) )); then
    echo "WARNING: CPU usage above 80%" >> $logfile
fi

if (( $(echo "$memory_usage > 80" | bc -l) )); then
    echo "WARNING: Memory usage above 80%" >> $logfile
fi

echo "-----------------------------------" >> $logfile
```

# Task 3 – User Account Management

## Objective

Create users from a list and generate random passwords.

---

## Script: create_users.sh

```bash
#!/bin/bash

input="user_list.txt"
output="credentials.txt"

while read username
do
    password=$(openssl rand -base64 8)

    sudo useradd "$username"
    echo "$username:$password" | sudo chpasswd

    echo "User: $username Password: $password" >> $output
done < "$input"

echo "Users created successfully."
```

Example `user_list.txt`

```
user1
user2
user3
```

# Task 4 – Automated Backup

## Objective

Compress a directory into a `.tar.gz` archive with the current date.

---

## Script: auto_backup.sh

```bash
#!/bin/bash

read -p "Enter directory path to backup: " dir

date=$(date +"%Y-%m-%d")
backup_file="backup_$date.tar.gz"

tar -czvf $backup_file $dir

echo "Backup saved as $backup_file"
```

# Task 5 – Simple To-Do List

## Objective

Create a simple command-line to-do list application.

---

## Script: todo.sh

```bash
#!/bin/bash

todo_file="todo.txt"

case "$1" in
add)
    shift
    echo "$@" >> $todo_file
    echo "Task added."
    ;;
view)
    cat $todo_file
    ;;
remove)
    sed -i "$2"d $todo_file
    echo "Task removed."
    ;;
*)
    echo "Usage:"
    echo "./todo.sh add 'task'"
    echo "./todo.sh view"
    echo "./todo.sh remove <line-number>"
    ;;
esac
```

Example Usage

```bash
./todo.sh add "Complete Bash assignment"
./todo.sh view
./todo.sh remove 1
```

# Task 6 – Automated Software Installation

## Objective

Install packages from a list and log installation results.

---

## Script: install_packages.sh

```bash
#!/bin/bash

logfile="installation.log"

while read package
do
    echo "Installing $package..." >> $logfile
    sudo apt install -y $package >> $logfile 2>&1

    if [ $? -eq 0 ]; then
        echo "$package installed successfully" >> $logfile
    else
        echo "$package installation failed" >> $logfile
    fi

done < packages.txt
```

Example `packages.txt`

```
git
curl
vim
htop
```

# Task 7 – Text File Processing

## Objective

Analyze a text file and extract useful information.

---

## Script: text_analysis.sh

```bash
#!/bin/bash

read -p "Enter file name: " file

echo "Line Count:"
wc -l $file

echo "Word Count:"
wc -w $file

echo "Character Count:"
wc -c $file

echo "Longest Word:"
tr -s '[:space:]' '\n' < $file | awk '{ if (length > max) { max = length; word = $0 } } END { print word }'
```
