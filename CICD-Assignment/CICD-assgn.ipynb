# Objective

The objective of this assignment is to design and implement a **complete CI/CD pipeline** that automatically builds, tests, scans for vulnerabilities, and deploys an application to a staging environment.

This project demonstrates real-world **DevOps and DevSecOps practices**, including automated testing, performance testing, security scanning, and deployment automation.

# Pipeline Overview

The CI/CD pipeline performs the following stages automatically:

| Stage | Description                               |
| ----- | ----------------------------------------- |
| 1     | Checkout repository                       |
| 2     | Setup Python environment                  |
| 3     | Install dependencies                      |
| 4     | Run unit tests                            |
| 5     | Run integration tests                     |
| 6     | Execute performance tests                 |
| 7     | Perform security scan                     |
| 8     | Build Docker image                        |
| 9     | Deploy application to staging environment |

This allows Jenkins to automatically pull the latest code and build the project.

## Steps Performed

1. Integrated OWASP ZAP into the CI pipeline.
2. Performed automated vulnerability scanning.
3. Checked for common security issues such as:

* Missing security headers
* Information disclosure
* Cross-site scripting risks
* insecure configurations

## Example Scan Command

```id="wq8kr1"
docker run --rm ghcr.io/zaproxy/zaproxy:stable zap-baseline.py -t http://127.0.0.1:5000
```

Security testing ensures the application is protected against common vulnerabilities.

---

📸 **Screenshot**



OWASP ZAP scan results showing detected warnings or passes.

---

# Task 7: Continuous Deployment (CD)

## Objective

To automatically deploy the application to a staging environment after successful pipeline execution.

## Tools Used

* Docker
* GitHub Actions

## Steps Performed

1. Created a Dockerfile for the application.
2. Built the Docker image within the CI pipeline.
3. Deployed the application container to a staging environment.

## Example Docker Commands

Build Image

```id="s9e51f"
docker build -t devops-api .
```

Run Container

```id="8v3qz4"
docker run -d devops-api
```

Verify Deployment

```id="3gkt3b"
docker ps
```

Continuous deployment ensures that the latest application version is automatically deployed after successful testing.

# CI/CD Pipeline Execution Flow

```
Code Push
   ↓
GitHub Actions Pipeline Triggered
   ↓
Install Dependencies
   ↓
Run Unit Tests
   ↓
Run Integration Tests
   ↓
Run Performance Tests
   ↓
Run Security Scan
   ↓
Build Docker Image
   ↓
Deploy to Staging Environment
```
